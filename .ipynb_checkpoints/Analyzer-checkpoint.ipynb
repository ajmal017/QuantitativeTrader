```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

time_interval='day'
asset=['TSLA']
```

```python
for symbol in asset:
    df = pd.read_csv(f'./Data/TSLA_day_intraday_trading.csv')
```

```python
df.columns = [col.strip() for col in df.columns]
df.columns = ['timestamp','open','high','low','close','volume']

df['volatility'] = df['high'] - df['low']
df['_hlMean'] =( df['high'] + df['low'])/2
df['vola_coeff'] = df['volatility'] / df['_hlMean']

df['day'] = [day[8:10] for day in df.timestamp]
df['month'] = [month[5:7] for month in df.timestamp]
df['year'] = [year[:4] for year in df.timestamp]
df['hour'] = [hour[-8:] for hour in df.timestamp]

time = df.timestamp
```

```python
fig, axs = plt.subplots(3)

fig.suptitle('TSLA:: (1) High (2) volatility (3) volatility coefficient ')
axs[0].plot(time, df['high'])
axs[1].plot(time, df['volatility'])
axs[2].plot(time,df['vola_coeff'])
```
